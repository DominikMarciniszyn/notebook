```python
sample_data = [
    {
        'name': 'sword',
        'type': 'item',
        'values': [{
            'attack': 10,
            'defense': 5
        }]
    },
    {
        'name': 'shield',
        'type': 'item',
        'values': [{
            'attack': 2,
            'defense': 10
        }]
    }
]
```

```python
print(sample_data)
```

```python
for item in sample_data:
    for key in item:
        if key == 'values':
            print(f"Attack value: {item['values'][0]['attack']}")
```

```python
for item in sample_data:
    print(f'Sample data type: {type(item)}')
```

```python
for item in sample_data:
    print(1)
    item['values'][0]['attack'] += 2

print(sample_data)
```
