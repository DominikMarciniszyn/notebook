```python
from datetime import time, date, datetime, timedelta
```

```python
print(f'Date of today: {date.today()}')
print(f'Datetime of today: {datetime.now()}')
```

```python
now = datetime.now()
dates = []

for index in range(0, 10):
    result = now + timedelta(seconds=15)
    now = result
    dates.append(result)
```

```python
for item in dates:
    print(f'The time is: {item}')
```
